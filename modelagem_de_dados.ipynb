```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_filmes = pd.read_csv('./database/desafio_indicium_imdb.csv', index_col=0)
```

```python
df_filmes.info()
```

```python
display(df_filmes)
```

```python
df_filmes['Gross'] = df_filmes['Gross'].str.replace(',', '', regex=False)
df_filmes['Gross'] = pd.to_numeric(df_filmes['Gross'], errors='coerce')

df_filmes['Meta_score'] = pd.to_numeric(df_filmes['Meta_score'], errors='coerce')

df_filmes['Released_Year'] = pd.to_numeric(df_filmes['Released_Year'], errors='coerce')
df_filmes['Released_Year'] = df_filmes['Released_Year'].astype('Int64')

df_filmes['Runtime'] = df_filmes['Runtime'].str.replace('min', '', regex=False).str.strip()
df_filmes['Runtime'] = pd.to_numeric(df_filmes['Runtime'], errors='coerce')
```

```python
df_filmes.isnull().sum()
```

```python
display(df_filmes)
```

```python
df_filmes.fillna({
    'Gross': df_filmes['Gross'].median(),
    'Meta_score': df_filmes['Meta_score'].mean(),
    'Certificate': 'Not Rated'
}, inplace=True)
```
